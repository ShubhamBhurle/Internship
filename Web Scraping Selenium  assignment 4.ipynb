# 1. Scrape the details of most viewed videos on YouTube from Wikipedia. Url = 
https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details: 
A) Rank 
B) Name 
C) Artist 
D) Upload date 
E) Views 

```python
import selenium

import pandas as pd

import time

from bs4 import BeautifulSoup

from selenium import webdriver

from selenium.common.exceptions import StaleElementReferenceException,NoSuchElementException

import requests

from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
#driver.maximize_window()
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
rank = []
video_names = []
video_uploader_name = []
views = []
upload_dates = []

try:
    rank_obj = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')[:30]
    for i in rank_obj:
        rank.append(i.text)
except NoSuchElementException :
    rank.append('-')
    
try:
    videoname_obj = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')[:30]
    for i in videoname_obj:
        video_names.append(i.text.split('[')[0][1:-1])
except NoSuchElementException :
    video_names.append('-')
    
try:
    uploader_obj = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')[:30]
    for i in uploader_obj:
        video_uploader_name.append(i.text)
except NoSuchElementException :
    video_uploader_name.append('-')
    
try:
    view_obj = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')[:30]
    for i in view_obj:
        views.append(i.text)
except NoSuchElementException :
    views.append('-')
    
try:
    date_obj = driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')[:30]
    for i in date_obj:
        upload_dates.append(i.text)
except NoSuchElementException :
    upload_dates.append('-')
```

```python
df=pd.DataFrame({'rank':rank, 'video_name':video_names,'uploader':video_uploader_name,'Views in Billions':views,'upload_date':upload_dates})
df
```

# 2. Scrape the details team India’s international fixtures from bcci.tv. 
Url = https://www.bcci.tv/.
You need to find following details: 
A) Match title (I.e. 1 ODI) 
B) Series 
C) Place 
D) Date 
E) Time 
Note: - From bcci.tv home page you have reach to the international fixture page through code. 

```python
driver = webdriver.Chrome()
driver.get("https://www.bcci.tv/")
```

```python
fixture_click = driver.find_element(By.XPATH,'(//a[@class="moreMatchesLink"])[1]')
fixture_click.click()
```

```python
Name=[]
Series=[]
Place=[]
Date=[]
Time=[]

try:
    name_tag = driver.find_elements(By.XPATH,'//span[@class="matchOrderText ng-binding ng-scope"]')
    for i in name_tag:
        Name.append(i.text)
except NoSuchElementException:
    Name.append('-') 
    
try:
    series_tag = driver.find_elements(By.XPATH,'//h5[@class="match-tournament-name ng-binding"]')
    for i in series_tag:
        Series.append(i.text)
except NoSuchElementException:
    Series.append('-')
    
try:
    place_tag = driver.find_elements(By.XPATH,'//span[@class="ng-binding ng-scope"]')
    for i in place_tag:
        Place.append(i.text)
except NoSuchElementException:
    Place.append('-')
    
try:
    date_tag = driver.find_elements(By.XPATH,'//div[@class="match-dates ng-binding"]')
    for i in date_tag:
        Date.append(i.text)
except NoSuchElementException:
    Date.append('-')

try:
    time_tag = driver.find_elements(By.XPATH,'//div[@class="match-time no-margin ng-binding"]')
    for i in time_tag:
        Time.append(i.text)
except NoSuchElementException:
    Time.append('-')
```

```python
print(len(Name),len(Series),len(Place),len(Date),len(Time))
```

```python
df=pd.DataFrame({'Match title':Name,' Series': Series,'Place':Place,'Date':Date,'Time':Time})
df
```

```python
Rank = []
State = []
Gsdp_19_20 = []
Gsdp_18_19 = []
Share_18_19 = []
Gdp_billion = []
```

```python
print(len(Rank),len(State),len(Gsdp_18_19),len(Gsdp_19_20),len(Share_18_19),len(Gdp_billion))
```

# 5. Scrape the details of top 100 songs on billiboard.com. Url = https:/www.billboard.com/
You have to find the following details: 
A) Song name 
B) Artist name 
C) Last week rank 
D) Peak rank 
E) Weeks on board 
Note: - From the home page you have to click on the charts option then hot 100-page link through code. 

```python
driver = webdriver.Chrome()
driver.get('https:/www.billboard.com/')
```

```python
charte = driver.find_element(By.XPATH,'/html/body/div[3]/header/div/div[2]/div/div/div[2]/div[2]/div/div/nav/ul/li[1]/a')
charte.click()
```

```python
hot_100 = driver.find_element(By.XPATH,'//a[@class="c-label__link lrv-u-color-brand-primary u-color-black@mobile-max a-font-primary-bold lrv-u-font-size-32 u-whitespace-nowrap@tablet"]')
hot_100.click()
```

```python
song_name = []
artist_name = []
last_week_rank = []
peak_rank = []
weeks_on_board = []
```

```python
song_tag = driver.find_elements(By.ID,"title-of-a-story")
for i in song_tag:
    song_name.append(i.text)
song_names = song_name[6::4][:100]

artist_tag = driver.find_elements(By.XPATH,'//span[@class="c-label  a-no-trucate a-font-primary-s lrv-u-font-size-14@mobile-max u-line-height-normal@mobile-max u-letter-spacing-0021 lrv-u-display-block a-truncate-ellipsis-2line u-max-width-330 u-max-width-230@tablet-only u-font-size-20@tablet"]')
for i in artist_tag:
    artist_name.append(i.text)
    
artist_tags = driver.find_elements(By.XPATH,'//span[@class="c-label  a-no-trucate a-font-primary-s lrv-u-font-size-14@mobile-max u-line-height-normal@mobile-max u-letter-spacing-0021 lrv-u-display-block a-truncate-ellipsis-2line u-max-width-330 u-max-width-230@tablet-only"]')
for i in artist_tags:
    artist_name.append(i.text)
    
week_rank_tag = driver.find_elements(By.XPATH,'//li[@class="o-chart-results-list__item // a-chart-color u-width-72 u-width-55@mobile-max u-width-55@tablet-only lrv-u-flex lrv-u-flex-shrink-0 lrv-u-align-items-center lrv-u-justify-content-center lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light u-background-color-white-064@mobile-max u-hidden@mobile-max"]')
for i in week_rank_tag:
    last_week_rank.append(i.text)
last_week_ranks=last_week_rank[0::2]

peak_rank_tag = driver.find_elements(By.XPATH,'//li[@class="o-chart-results-list__item // a-chart-bg-color a-chart-color u-width-72 u-width-55@mobile-max u-width-55@tablet-only lrv-u-flex lrv-u-flex-shrink-0 lrv-u-align-items-center lrv-u-justify-content-center lrv-u-background-color-grey-lightest lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light u-hidden@mobile-max"]')
for i in peak_rank_tag:
    peak_rank.append(i.text)
peak_ranks=peak_rank[1::2]

week_on_board_tag = driver.find_elements(By.XPATH,'//li[@class="o-chart-results-list__item // a-chart-color u-width-72 u-width-55@mobile-max u-width-55@tablet-only lrv-u-flex lrv-u-flex-shrink-0 lrv-u-align-items-center lrv-u-justify-content-center lrv-u-border-b-1 u-border-b-0@mobile-max lrv-u-border-color-grey-light u-background-color-white-064@mobile-max u-hidden@mobile-max"]')
for i in week_on_board_tag:
    weeks_on_board.append(i.text)
    weeks_on_boards = weeks_on_board[1::2]
```
